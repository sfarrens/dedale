```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import galsim
```

```python
postage = [41, 41]
pixel_size = 0.05
hlr = 0.3
psf_size = 0.0

phi = galsim.Angle(0.0, galsim.degrees)

# mono_gal = galsim.InclinedExponential(half_light_radius=hlr, inclination=phi)
mono_gal = galsim.Exponential(half_light_radius=hlr)
mono_gal = galsim.Convolve([mono_gal, galsim.Gaussian(fwhm=psf_size)])
# mono_gal = mono_gal.shear(q=0.5, beta=galsim.Angle(45.0, galsim.degrees))
mono_gal = mono_gal.shear(e1=0.1, e2=0.2)

image = galsim.ImageF(*postage)               
# mono_gal.drawImage(image=image, scale=pixel_size)
mono_gal.drawImage(image=image, scale=pixel_size, method='no_pixel')
data = image._image.array

# data = mono_gal.draw(scale=pixel_size).array
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(data)
plt.colorbar()
plt.show()
```
